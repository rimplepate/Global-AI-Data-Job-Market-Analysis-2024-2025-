```python
file_path = "ai_job_dataset.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['salary_usd'], bins=50)
plt.title("Salary Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='experience_level', y='salary_usd', data=df)
plt.title("Salary by Experience Level")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
skills_series = df['required_skills'].str.split(',').explode()
top_skills = skills_series.value_counts().head(15)

top_skills.plot(kind='bar', figsize=(10,5))
plt.title("Top 15 In-Demand Skills")
plt.show()
```
